```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
school_df = pd.read_csv('/content/School_Details.csv')
# make sure it loaded correctly
school_df.head(2)
```

```python
#create sqlite engine for a database called schools
engine=create_engine('sqlite:///schools.db', echo=True)
#create connection to engine
conn=engine.connect()
```

```python
sqlite_table='school_details'
school_df.to_sql(sqlite_table, conn,if_exists='fail')
```

```python
sql="""
SELECT DistrictName
FROM school_details
Limit3;
"""

pd.read_sql_query(sql, engine)
```

##Groupby
Use with aggregate functions
Use AS to name results of GROUPBY aggregate functions

```python
sql="""SELECT LocatedinRuralOrUrban, AVG (`DistrictHQDistance (in Km)`)
FROM school_details
GROUP BY LocatedInRuralOrUrban;
"""
pd.read_sql_query(sql, engine)
```

```python
sql="""SELECT PostOfficeName, MAX(`DistrictHQDistance (in Km)`)
FROM school_details
GROUP BY PostOfficeName;
"""

pd.read_sql_query(sql,engine)
```

```python
sql="""SELECT CRCName, COUNT(*)
FROM school_details
GROUP BY CRCName
ORDER BY COUNT(*) DESC;
"""
pd.read_sql_query(sql, engine)
```

```python
sql="""SELECT CRCName, COUNT(*) AS NumberOfSchools
FROM school_details
GROUP BY CRCName
ORDER BY NumberofSchools DESC;
"""
pd.read_sql_query(sql,engine)
```

##HAVING
filter values based on the results of aggregate functions with GROUP BY

similar to WHERE but HAVING must be used with GROUPBY

```python
sql="""SELECT CRCName, COUNT(*)
FROM school_details
GROUP BY CRCName
HAVING COUNT (*)>20;
"""
pd.read_sql_query(sql,engine)
```

```python
sql="""SELECT CRCName, COUNT(*) AS NumberofSchools
FROM school_details
GROUP BY CRCName
HAVING NumberofSchools>20;
"""
pd.read_sql_query(sql,engine)
```

```python
sql="""SELECT SubdivisionName, AVG(`DISTRICTHQDistance (in Km)`) AS Avg_HQ_Distance
FROM school_details
GROUP BY SubdivisionName
HAVING Avg_HQ_Distance>30;
"""
pd.read_sql_query(sql, engine)
```
